*   텍스트 마이닝 (Text Mining) 이란?
- 비/반정형 텍스트 데이터에서 자연어처리(Natural Language Processing)기술에 기반하여 유용한 정보를 추출, 가공하는 것을 목적으로 하는 기술이다.

 

> 텍스트 마이닝의 주요 기술

- 자연어 처리 (파싱, 형태소 분석, 품사 태깅, 관계 추출, 의미 추출)

- 언어모델링 (언어 감지, 규칙기반 개체명과 상용어 인식)

- 기계 학습 알고리즘 (반복훈련을 통해 습득한 정보 사용능력을 개선)

- 마이닝 기술 (각종 통계적 기법을 활용한 정보 분류 및 분석 기술)

 

 

텍스트 마이닝의 3단계 & 각 단계에서 필요한 준비
1. 문서 수집 (Crawling)

2. 형태소 분석 (NLP) (Konlpy)

3. 시각화 (Word Cloud)

2. 형태소 분석 (Konlpy)

  1) 형태소 분석 개요

     문서(document) > 문단(paragraph) > 문장(sentence) > 어절 > 형태소 > 음절

      * 형태소 : 의미를 가진 가장 작은 말의 단위. 더 나누면 뜻을 잃어버림.

  2) 형태소 분석 관련 패키지 설치

```python
import konlpy
```

```python
from konlpy.corpus import kolaw
c = kolaw.open('constitution.txt').read()
#예제파일 가져오기
```

```python
print(type(c)) #가져온 데이터의 type을 확인
print(len(c))
print(c[:600])
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
c_sent = sent_tokenize(c)
print(len(c_sent))
print(c_sent[:5])

#nltk로 tokenize 시도
```

```python

from nltk.tokenize import word_tokenize
c_words = word_tokenize(c)
len(c_words)

#word로 tokenize
```

```python
" ".join(c_words[:50])
```

```python
from konlpy.tag import Okt
okt = Okt()
tokens_c = okt.morphs(c) 

#형태소 단위로 tokenize
```

```python
len(tokens_c)
```

```python
type(tokens_c)
```

```python
" ".join(tokens_c[:50])
# 형태소 단위로 분리한 tokens를 join
```

```python
import nltk
c_nltk_text = nltk.Text(tokens_c, name = "대한민국헌법") 
c_nltk_text

#tokens_c 대신 c 사용
```

```python
# 모든 단어의 수
print(len(c_nltk_text.tokens))
# 서로 다른 단어의 수
print(len(set(c_nltk_text.tokens)))
```

```python
#단어와 단어의 빈도를 dictionary 형태로 표현 
c_nltk_text.vocab()
```

```python
#상위 빈도 단어들의 빈도를 그래프로 출력
%matplotlib inline
c_nltk_text.plot(20)
```

```python
tokens_new = []
for token in tokens_c:
    if len(token.strip()) > 1: 
        tokens_new.append(token.strip())
print(len(tokens_new))

# 스페이스를 제외한 나머지 문자열의 길이가 둘 이상인 단어만 포함하여 개수 계산
```

```python
c_nltk_text = nltk.Text(tokens_new, name = "대한민국헌법")
c_nltk_text.plot(20)

# 그래프로 재출력
```

```python
#특정한 단어의 빈도 측정('대한민국')
c_nltk_text.count('대한민국')
```

```python
# 단어들 중 주어진 단어들의 위치를 표시
c_nltk_text.dispersion_plot(['대한민국', '민주', '헌법'])
```

```python
c_nltk_text.concordance('헌법')
# concordance를 이용해 '헌법' 이라는 단어 앞뒤 단어들을 보여줌
```

```python
nltk.download('stopwords')
# 함께 많이 나타난 단어들을 출력
c_nltk_text.collocations()
```
